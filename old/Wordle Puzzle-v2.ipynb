```python
def same_subword(worda, wordb, comb):
    for idx in comb:
        if worda[idx] != wordb[idx]:
            return False
    return True

max_count = 0
max_word = ""
max_comb = []

all_count = []
all_word = []
all_comb = []


for word_i in word_list:
    for comb in combs:
        count = 0
        for word_j in word_list:
            if same_subword(word_i, word_j, comb):
                count += 1
             
        if count > 5:
            all_count.append(count)
            all_word.append(word_i)
            all_comb.append(comb)
        
        if (count > max_count):
            max_count = count
            max_comb = comb
            max_word = word_i
```

```python
def filter_word(word):
    cond0 = True
    cond1 =  word[1] == "O" #True
    cond2 =  word[2] == "W" #True
    cond3 = word[3] == 'E'
    cond4 = word[4] == 'R'
    return (cond0 and cond1 and cond2 and cond3 and cond4)

filtered_words = []
for word in word_list:
    if filter_word(word):
        filtered_words.append(word)
```

```python
seq_eg = "ZONAL->QUASH->AFFIX->MAMMA->WAVER->RARER->RACER->EATER->EAGER->BAKER->PARER->PAPER->PAYER"

def seq_analysis(seq_str):
    alphabet = [c for c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"]
    seq = [word for word in seq_str.split('->')]
    unused = alphabet
    for word in seq:
        for char in word:
            if char in alphabet:
                unused.remove(char)
    return unused

seq_analysis(seq_eg)
```
